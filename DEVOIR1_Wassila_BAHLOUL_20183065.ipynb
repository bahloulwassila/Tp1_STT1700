A MODIFIER
Les paramètres statistiques $\bar{x}$ (la moyenne) et $s_{\sigma}$ (l'écart-type) sont appropriés pour résumer des distributions qui suivent une loi normale ou qui présentent une forme approximative de cloche. La moyenne $\bar{x}$ représente le point central de la distribution, tandis que l'écart-type \$s_{\sigma}$ mesure la dispersion des données autour de cette moyenne.

Il est toutefois crucial de souligner que ces statistiques sont sensibles aux valeurs extrêmes. En présence de données non normalement distribuées, d'autres mesures de centralité et de dispersion peuvent être plus appropriées. Par exemple, la médiane $\tilde{x}$ et l'écart interquartile (IQR) sont des mesures robustes qui ne sont pas influencées par les valeurs extrêmes, ce qui les rend préférables dans le cas de distributions asymétriques ou à queues épaisses.

En conclusion, $\bar{x}$ et \$s_{\sigma}$ sont des choix pertinents pour des distributions normales ou approximativement normales, mais il convient de considérer la nature spécifique des données pour déterminer les mesures de résumé les plus appropriées dans d'autres contextes.

```r
DureeFemme = c(180, 120, 180, 360, 240, 120, 180, 120, 240, 170,
                  150, 120, 180, 180, 150, 200, 150, 180, 150, 180,
                  120, 60, 120, 180, 180, 90, 240, 180, 115, 120)
  
  
q1_femmes <- quantile(DureeFemme, 0.25)
print(q1_femmes)

q3_femmes <- quantile(DureeFemme, 0.75)
print(q3_femmes)

iqr_femmes <- q3_femmes-q1_femmes
print(iqr_femmes)

limite_inf_femmes <- q1_femmes - 1.5*iqr_femmes
print(paste("Limite inférieure:", limite_inf_femmes))

limite_sup_femmes <- q3_femmes + 1.5* iqr_femmes
print(paste("Limite supérieure:", limite_sup_femmes))

valeurs_aberrantes_femmes <- DureeFemme[DureeFemme < limite_inf_femmes | DureeFemme > limite_sup_femmes]
print(valeurs_aberrantes_femmes)
```

```r
DureeHomme = c(90, 120, 30, 90, 200, 90, 45, 30, 120, 75, 150,
                  120, 60, 240, 300, 240, 60, 120, 60, 30, 30, 230,
                  120, 95, 150, 0, 200, 120, 120, 180)
  
  
q1_hommes <- quantile(DureeHomme, 0.25)
print(q1_hommes)

q3_hommes <- quantile(DureeHomme, 0.75)
print(q3_hommes)

iqr_hommes <- q3_hommes-q1_hommes
print(iqr_hommes)

limite_inf_hommes <- q1_hommes - 1.5*iqr_hommes
print(paste("Limite inférieure:", limite_inf_hommes))

limite_sup_hommes <- q3_hommes + 1.5* iqr_hommes
print(paste("Limite supérieure:", limite_sup_hommes))

valeurs_aberrantes_hommes <- DureeHomme[DureeHomme < limite_inf_hommes | DureeHomme > limite_sup_hommes]
print(valeurs_aberrantes_hommes)
```
